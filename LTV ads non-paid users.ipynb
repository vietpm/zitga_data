```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel("data/ltv ads.xlsx", parse_dates = ["install_date"], index_col = 0)
df.head()
```

```python
fig, ax = plt.subplots(3, 2, figsize = (15, 12))
df.ltv1.plot(ax = ax[0][0])
df.ltv3.plot(ax = ax[0][1])
df.ltv7.plot(ax = ax[1][0])
df.ltv30.plot(ax = ax[1][1])
df.ltv60.plot(ax = ax[2][0])
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (20, 12))
# month_list = df.index.month.unique().sort_values()
month_list = [4, 5, 6, 7]
for i in month_list:
    plt.plot(df["2019-"+ str(i)].index.day, df["2019-"+ str(i)].ltv1.values, label = i)
ax.legend()
```
