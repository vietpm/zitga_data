# Import libraries and load dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
```

```python
raw = pd.read_excel("sk install.xlsx", sheet_name = "Sheet")
raw.info()
```

```python
df = raw.pivot_table(index = ["event_date", "country_code"], columns = "media_source", values = "install").reset_index().fillna(0)
df.to_excel("sk install pv.xlsx")
```

```python
source_list = list(df.columns)
source_list.remove("event_date")
source_list.remove("country_code")
source_list.remove("Organic")
source_list
```

```python
df["Paid"] = df[source_list].sum(axis = 1)
df
```

```python
df["Other"] = df["Paid"] - df["Facebook Ads"] - df.googleadwords_int
df
```

```python
# df.to_excel("sk install pv.xlsx")
```

```python
KR_date = df[df.event_date >= '2020-07-04'].event_date.unique()
US_date = df[(df.event_date >= '2020-06-14') & (df.event_date != '2020-06-17')].event_date.unique()
print(KR_date)
print(US_date)
```

# Linear Regression

```python
# cross validate (train)
# KR
country_code = ["KR", "US"]
date = [KR_date, US_date]

for idx, c in enumerate(country_code):
    reg1 = LinearRegression()
    X = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])), ["Facebook Ads", "googleadwords_int", "Other"]]
#     X = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])), ["Facebook Ads", "googleadwords_int"]]
    y = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])),"Organic"]
    cv_results = cross_validate(reg1, X, y, cv = 10, return_train_score = True, return_estimator = True)
    print("---------------%s--------------" % c)
    for i, model in enumerate(cv_results["estimator"]):
        print("- Lần chạy %d:" % (i+1))
        print("R squared: %f" % cv_results["train_score"][i])
        print("K factor: %s" % model.coef_)
        print("Base organic: %f" % model.intercept_)
    print("Mean train score: %f" % cv_results["train_score"].mean())
    print("Std train score: %f" % cv_results["train_score"].std())
    print("")
```

```python
# train all
country_code = ["KR", "US"]
for idx, c in enumerate(country_code):
    reg1 = LinearRegression()
    X = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])), ["Facebook Ads", "googleadwords_int", "Other"]]
#     X = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])), ["Facebook Ads", "googleadwords_int"]]
    y = df.loc[(df.country_code == c) & (df.event_date.isin(date[idx])),"Organic"]
    reg1.fit(X, y)
    print("----------%s----------" % c)
    print("R squared: %f" % reg1.score(X, y))
    print("Organic = Facebook * %f + Google * %f + %f" % (reg1.coef_[0], reg1.coef_[1], reg1.intercept_))
    print("")
```
